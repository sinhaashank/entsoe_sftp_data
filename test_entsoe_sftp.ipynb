```python
###########################################################################################
def filter_data_by_country(
    df = None,
    df_coutries_bidding_zones = None,
    year = None,
    month = None,
    country_code = None,
    column_AreaName = 'AreaName',
    column_MapCode = 'MapCode',
):
    from datetime import datetime
    import pandas as pd

    date = datetime(year=year, month=month, day=1)

    area_name, map_code = tuple(
        df_coutries_bidding_zones
        .loc[df_coutries_bidding_zones['CountryCode'] == coutry_code]
        .set_index('change date')
        .sort_index()
        .asof(date) # returns series
        [['AreaName', 'MapCode']].tolist()
    )
    
    df_focus = pd.DataFrame()
    
    #-------------------------------------------------------------------------------------
    df_focus = df_coutries_bidding_zones.loc[
        (df_coutries_bidding_zones[column_AreaName] == area_name)
        & (df_coutries_bidding_zones[column_MapCode] == map_code)
    ]
    
    return df_focus
```
